**Ejercicio de aplicaci√≥n**

\begin{equation*}
    \begin{cases}
    x^2 - 10x + y^2 = -8 \\
    xy^2 + x - 10y = -8
    \end{cases}
\end{equation*}



ejemplo.jpg

```python

import numpy as np

def f(x):
    # Sistema de ecuaciones no lineales
    return np.array([x[0]**2 - 10*x[0] + x[1]**2 + 8,
                     x[0] * x[1]**2 + x[0]-10*x[1] + 8])


def jacobian(x):
    # Matriz Jacobiana
    return np.array([
        [2*x[0] - 10, 2*x[1]],
        [x[1]**2 + 1, 2*x[0]*x[1] - 10]
    ])


def newton_method(x0, tol=1e-4, max_iter=100):
    # M√©todo de Newton para sistemas no lineales
    x = np.array(x0, dtype=float)

    print("\n----- Inicio del M√©todo de Newton -----\n")
    # Imprimir encabezado de la tabla
    print(f"{'Iteraci√≥n':<10}{'x':<15}{'y':<15}{'Cambio Relativo':<20}")
    print("="*60)

    for i in range(max_iter):
        J = jacobian(x)  # Jacobiana
        F = f(x)  # Evaluar funciones
        delta_x = np.linalg.solve(J, -F)  # Resolver J * delta_x = -F
        x += delta_x  # Actualizar soluci√≥n
        print(f"{i+1:<10}{x[0]:<15.6f}{x[1]:<15.6f}{np.linalg.norm(delta_x):<.6f}")
        if np.linalg.norm(delta_x) < tol:
            print(f"Convergi√≥ en {i+1} iteraciones.")
            return x
    print("No convergi√≥.")
    return x


# Estimaci√≥n inicial para el punto x0 = [0,0]
x0 = [0,0]
sol = newton_method(x0)
print("Soluci√≥n aproximada:", sol)

# Estimaci√≥n inicial para el punto x0 = [2,2]
x0 = [2,2]
sol = newton_method(x0)
print("Soluci√≥n aproximada:", sol)
```

**Tarea en aula**

\begin{equation*}
    \begin{cases}
    x^2 + y = 4 \\
    x + y^2 = 4
    \end{cases}
\end{equation*}

ejemplo.jpg

```python
# Modificar en f(x), jacobian(x) para hacer los cambios y la estimacion inicial x0

def f(x):
    # Sistema de ecuaciones no lineales
    return np.array([x[0]**2 + x[1] -4,
                     x[0] + x[1]**2 -4])


def jacobian(x):
    # Matriz Jacobiana
    return np.array([
        [2*x[0], 1],
        [1, 2*x[1]]
    ])


def newton_method(x0, tol=1e-4, max_iter=100):
    # M√©todo de Newton para sistemas no lineales
    x = np.array(x0, dtype=float)

    print("\n----- Inicio del M√©todo de Newton -----\n")
    # Imprimir encabezado de la tabla
    print(f"{'Iteraci√≥n':<10}{'x':<15}{'y':<15}{'Cambio Relativo':<20}")
    print("="*60)

    for i in range(max_iter):
        J = jacobian(x)  # Jacobiana
        F = f(x)  # Evaluar funciones
        delta_x = np.linalg.solve(J, -F)  # Resolver J * delta_x = -F
        x += delta_x  # Actualizar soluci√≥n
        print(f"{i+1:<10}{x[0]:<15.6f}{x[1]:<15.6f}{np.linalg.norm(delta_x):<.6f}")
        if np.linalg.norm(delta_x) < tol:
            print(f"Convergi√≥ en {i+1} iteraciones.")
            return x
    print("No convergi√≥.")
    return x

# Estimaci√≥n inicial para el punto x0 = [1, 1]
x0 = [1, 1]
sol = newton_method(x0)
print("Soluci√≥n aproximada:", sol)

# Estimaci√≥n inicial para el punto x0 = [1, -1]
x0 = [1, -1]
sol = newton_method(x0)
print("Soluci√≥n aproximada:", sol)

# Estimaci√≥n inicial para el punto x0 = [-1, 1]
x0 = [-1, 1]
sol = newton_method(x0)
print("Soluci√≥n aproximada:", sol)

# Estimaci√≥n inicial para el punto x0 = [-1, -1]
x0 = [-1, -1]
sol = newton_method(x0)
print("Soluci√≥n aproximada:", sol)
```
